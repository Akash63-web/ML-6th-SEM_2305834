```python
x=[-7,-5,-3,-1,1,3,5,7]
y=[80,90,92,83,94,99,92,104]
```

```python
def mean(values):
    return sum(values)/len(values)
```

```python
def linear_regression(x,y):
    x_mean=mean(x)
    y_mean=mean(y)

    num= sum((x[i]-x_mean) * (y[i]-y_mean) for i in range(len(y)))
    deno= sum((x[i]-x_mean)**2 for i in range(len(x)))

    m=num/deno
    b=y_mean-m*x_mean

    return m,b
```

```python
def predict(x,m,b):
    return m*x + b
```

```python
#Train the model
m,b=linear_regression(x,y)
```

```python
print("Slope: ", m)
print("Intercept (b): ", b)
```

```python
#Prediction time
print("Prediction for x=9: ",predict(9,m,b))
print("Prediction for x=10: ",predict(10,m,b))
```
