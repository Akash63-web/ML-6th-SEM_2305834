```python
import requests
import pandas as pd
```

```python
url = "https://openlibrary.org/search.json?q=python"  
```

```python
response=requests.get(url)
data=response.json()
```

```python
for i in range(20):
    print(data["docs"][i]["title"])
```

```python
titles = [data["docs"][i]["title"] for i in range(30)] 
```

```python
df = pd.DataFrame({"title": titles})
df.to_csv("books.csv", index=False)
```
